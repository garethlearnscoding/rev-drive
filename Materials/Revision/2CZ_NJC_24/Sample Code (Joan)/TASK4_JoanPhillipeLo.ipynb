## Task 4.1
Write a Python program that uses SQL code to create the database <span style="font-family: Courier New,monospace;">ESPORTS</span> with the two tables given. Define the primary and foreing keys for each table.

<div style="text-align:right;">[4]</div>

Run your program to test the database has been set up correctly and save your database as `esports.db`.

<div style="text-align:right;">[1]</div>

```python
import sqlite3

conn = sqlite3.connect('esports.db')
cursor = conn.cursor()
cursor.execute("""
    CREATE TABLE IF NOT EXISTS people (
        PersonID INTEGER PRIMARY KEY AUTOINCREMENT,
        FullName TEXT NOT NULL, 
        DateofBirth TEXT NOT NULL,
        IsPlayer INTEGER, 
        IsStaff INTEGER
    )
""")
cursor.execute("""
    CREATE TABLE IF NOT EXISTS player (
        PersonID INTEGER PRIMARY KEY AUTOINCREMENT,
        TeamName TEXT NOT NULL,
        CharacterName TEXT NOT NULL,
        EventName INTEGER,
        Score INTEGER,
        FOREIGN KEY (PersonID) REFERENCES people(PersonID)
    )

""")
conn.commit()
conn.close()
```

```python
import string 
class Person:
    def __init__(self, full_name: str, date_of_birth:str):
        self.full_name = full_name
        self.date_of_birth = date_of_birth

    def is_player(self):
        return 'Maybe'
    def is_staff(self):
        return 'Maybe'
    def event_name(self):
        fullname = self.full_name.replace(' ', '').translate(str.maketrans('','', string.punctuation))
        user = fullname + self.date_of_birth[5:7]+ self.date_of_birth[8:10]
        return user

class Player(Person):
    def __init__(self, full_name:str, date_of_birth:str, char_name:str, team_name:str, score:int):
        super().__init__(full_name, date_of_birth)
        self.char_name = char_name
        self.team_name = team_name
        self.score = score
    
    def event_name(self):
        user = self.char_name + ' <' + self.team_name + '>'
        return user
    
    def is_player(self):
        return True

class Staff(Person):
    def __init__(self, full_name: str, date_of_birth:str):
        super().__init__(full_name, date_of_birth)

    def is_staff(self):
        return True
    
    def event_name(self):
        fullname = self.full_name.replace(' ', '').translate(str.maketrans('','', string.punctuation))
        user = fullname + self.date_of_birth[5:7]+ self.date_of_birth[8:10] + ' ' + 'Staff'
        return user  
```

## Task 4.3
The text file, `TASK4PEOPLE.CSV`, contains data items for a number of people. Each data item is separated by a commn, with each person's data on a newline as follows:
* full name
* date of birth in the form YYYY-MM-DD
* a string indicating whether the person is `Staff`,`Player` or  `Person`.
<br>

Another text file, `TASK4PLAYERS.CSV`, contains additional data item for the people who are also players. Each data time is separated by a comma, with each peron's data on a newline as follows:

* full name
* date of birth in the form YYYY-MM-DD
* team name
* character name
* competition score

Write program code to read in the information from the text files, `TASK4PEOPLE.CSV` and `TASK4PLAYERS.CSV` creating an instance of the appropriate class for each person (either `Player`,`Staff` or `Person`). Store all this instances in list named `Event_People`.

<div style="text-align:right;">[6]</div>

Write program code to insert all information from the text files, `TASK4PEOPLE.CSV` and `TASK4PLAYERS.CSV` into the two tables in `esports.db` database. <br>

Run the program.

<div style="text-align:right;">[8]</div>


```python
import csv, sqlite3
with open('./Resources/TASK4/TASK4PEOPLE.CSV') as f:
    data = [record for record in csv.reader(f)]

with open('./Resources/TASK4/TASK4PLAYERS.CSV') as f:
    records = [record for record in csv.reader(f)]

people_data = {}
players_data = {}

for record in data:
    name, dob, status = record[0], record[1], record[2]
    people_data[name] = dob, status

for record in records:
    name, dob, team, char, score = record[0], record[1], record[2], record[3], record[4]
    players_data[name] = (dob, team, char, int(score))

result = []

for name, (dob, status) in people_data.items():
    if status == "Player" and name in players_data:
        dob,team,char,score = players_data[name]
        result.append(Player(name,dob,char,team,score))
    elif status == "Staff":
        result.append(Staff(name,dob))
    else:
        result.append(Person(name,dob))


conn = sqlite3.connect("esports.db")
cursor = conn.cursor()

for person in result:
    if isinstance(person, Player):
        cursor.execute("INSERT INTO people (FullName, DateOfBirth, IsPlayer, IsStaff) VALUES (?,?,?,?)", 
        (person.full_name, person.date_of_birth, 1, 0)    
        )
        cursor.execute("INSERT INTO player (TeamName, CharacterName, EventName, Score) VALUES (?,?,?,?)", 
        (person.team_name, person.char_name, person.event_name(), person.score))
    elif isinstance(person, Staff):
        cursor.execute("INSERT INTO people (FullName, DateOfBirth, IsPlayer, IsStaff) VALUES (?,?,?,?)",
        (person.full_name, person.date_of_birth, 0, 1))
    else:
        cursor.execute("INSERT INTO people (FullName, DateOfBirth, IsPlayer, IsStaff) VALUES (?,?,?,?)",
        (person.full_name, person.date_of_birth, 0, 0))
    
conn.commit()
conn.close()
```
